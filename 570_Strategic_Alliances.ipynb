## Load in dataset or Generate the Synthetic dataset

```python
# Set random seed for reproducibility
np.random.seed(12)

# Generate a 20x20 matrix with random 0s and 1s
matrix_size = 20
synthetic_matrix = np.random.randint(2, size=(matrix_size, matrix_size))

# Make the matrix symmetric (ties are bidirectional)
synthetic_matrix = np.triu(synthetic_matrix) + np.triu(synthetic_matrix, 1).T

# Set the diagonal to 0 (no self-loops)
np.fill_diagonal(synthetic_matrix, 0)

# Create company labels
companies = [f'Company_{i+1}' for i in range(matrix_size)]

# Convert to DataFrame
financial_ties_df = pd.DataFrame(synthetic_matrix, index=companies, columns=companies)

# Save the matrix to an Excel file
#financial_ties_df.to_excel('FinancialTies_Synthetic.xlsx')

# Display the DataFrame
print(financial_ties_df.head())
```

### 1. Create a matrix representation of the network.

```python
# Load the synthetic dataset from the Excel file
financial_ties_df = pd.read_excel('FinancialTies_Synthetic.xlsx', index_col=0)

# Create a NetworkX graph from the adjacency matrix
G = nx.from_pandas_adjacency(financial_ties_df)

# Print graph information
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
density = nx.density(G)

print(f"Graph with {num_nodes} nodes and {num_edges} edges")
print(f"Density of the Network: {density:.4f}")

# Degree of each node
print("\nDegree of each node:")
for node, degree in G.degree():
    print(f'{node}: Degree {degree}')

# Check if the graph is connected
is_connected = nx.is_connected(G)
print(f"\nIs the graph connected? {is_connected}")
```

The graph shows 20 companies with 96 connections and a density of 0.5053, indicating a moderately connected network. Company_5 has the highest degree (14), suggesting strong integration. The graph is connected, meaning there are no isolated entities, allowing for potential influence throughout the network.

### 6. Analyze the matrix to identify the impact of interlocking boards.

Interlocking boards occur when companies share board members or influential decision-makers, potentially aligning their strategies, forming alliances, and influencing competitive actions. We can use use centrality metrics to identify potential interlocking boards.

Companies with high degree centrality have many direct ties, indicating potential shared board members or high influence through connections.

```python
# Identify top companies by degree centrality (potential interlocking boards)
top_companies_by_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top Companies with Potential Interlocking Boards (by Degree Centrality):")
for company, centrality in top_companies_by_degree:
    print(f"{company}: {centrality:.4f}")
```

Company_5 and Company_2 are highly connected, suggesting they may have overlapping board members or serve as influential nodes in the network.

Companies with high betweenness centrality bridge separate clusters, influencing how information and decisions flow across groups.

```python
# Identify top companies by betweenness centrality (influencing multiple groups)
top_companies_by_betweenness = sorted(nx.betweenness_centrality(G).items(), key=lambda x: x[1], reverse=True)[:5]
print("\nTop Companies with Cross-Company Influence (by Betweenness Centrality):")
for company, centrality in top_companies_by_betweenness:
    print(f"{company}: {centrality:.4f}")
```

Company_6 and Company_5 act as intermediaries between groups, increasing their strategic influence. These companies are critical in connecting otherwise disconnected clusters, enabling or preventing alliances.

**Impact of Interlocking Boards**  

Interlocking boards facilitate strategic alliances by aligning the interests of companies through shared board members, making it easier to establish partnerships or joint ventures. However, these overlapping decision-makers also present a risk of collusion, as they may engage in anti-competitive behavior that unfairly influences the market. Additionally, interlocking boards allow companies to shape industry trends by coordinating strategic actions, such as mergers, which provide connected companies with a competitive edge.  

Bipartite graphs offer another useful method for detecting interlocking boards by modeling relationships between companies and shared board members. In this structure, companies and board members form two distinct node sets, with edges representing board memberships. Projecting the bipartite graph onto a unipartite graph reveals companies that share board members, with edge weights indicating the strength of their connections. This approach complements centrality measures by uncovering governance overlaps that can influence alliances, market dynamics, and competition.

### 9. Detect Structural Holes and Explain Their Impact on Social Closures  

#### Definition of Structural Holes  
A structural hole refers to a gap between companies (nodes) that are not directly connected but could benefit from collaboration. Bridging these holes allows access to non-redundant information and strategic advantages. Social closure occurs when such gaps are filled, promoting trust and cooperation, but it may also reduce innovation by limiting access to diverse partnerships.

#### Identifying Structural Holes and Key Bridges  
Structural holes can be detected by identifying companies with **high betweenness centrality**, as these companies act as intermediaries between clusters. 

```python
# Calculate betweenness centrality for all companies
betweenness_centrality = nx.betweenness_centrality(G)

# Identify companies with high betweenness centrality (bridging structural holes)
structural_hole_bridges = [
    company for company, centrality in betweenness_centrality.items() 
    if centrality > 0.07  # Adjust threshold as needed based on output
]
print("Companies Bridging Structural Holes:")
print(structural_hole_bridges)

# Use community detection to find isolated groups (structural holes)
from networkx.algorithms.community import girvan_newman

communities = next(girvan_newman(G))
print("\nDetected Communities:")
for i, community in enumerate(communities):
    print(f"Community {i + 1}: {list(community)}")

# Identify isolated nodes or small clusters (structural holes)
isolated_communities = [community for community in communities if len(community) < 3]
print("\nIsolated or Small Communities (Potential Structural Holes):")
print(isolated_communities)
```
