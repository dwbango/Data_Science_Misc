```python
_='''
Uses numpy to generate random stock returns following a normal distribution.
Plots a histogram of these returns using matplotlib.
Overlays the theoretical PDF of the normal distribution on top of the histogram.
Calculates the 95% VaR using the inverse of the CDF (percent-point function).
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulated stock returns (mean=0.001, std deviation=0.02)
mean = 0.01
std_dev = 0.02
returns = np.random.normal(mean, std_dev, 1000)

# Plotting histogram of returns
plt.hist(returns, bins=50, density=True, alpha=0.6, color='g')

# Overlaying the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

plt.title("Stock Returns Distribution with Fitted PDF")
plt.show()

# Calculating 95% VaR using inverse of the CDF
var_95 = mean - norm.ppf(0.95, loc=mean, scale=std_dev)*std_dev
print("Value at Risk (95% confidence):", var_95)
```
