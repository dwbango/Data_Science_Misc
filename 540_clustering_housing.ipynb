## Feature Engineering

```python
# One-hot encode categorical variable 'ocean_proximity'
data_encoded = pd.get_dummies(data, columns=['ocean_proximity'])
```

```python
# Scale numerical variables
scaler = StandardScaler()
numerical_cols = [col for col in data_encoded.columns if col != 'ocean_proximity']
data_scaled = scaler.fit_transform(data_encoded[numerical_cols])
data_scaled = pd.DataFrame(data_scaled, columns=numerical_cols)
```

```python
# Combine scaled numerical variables with encoded categorical variables (preprocessed)
data_preprocessed = pd.concat([data_scaled, data_encoded.drop(columns=numerical_cols)], axis=1)
```
