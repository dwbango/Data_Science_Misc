```python
_='''
Suppose you are given a biased coin with unknown probability of landing heads. 
Design an experiment to estimate the probability of heads using probability theory.
How would you determine the number of coin flips to achieve a certain level of confidence 
in your estimate? Additionally, how can Python be used to simulate this experiment and 
visualize the results?
'''
```

```python
_='''
1. Experiment Design:
- Coin Flipping: Flip the coin 'n' times.
- Counting: Determine the number of times it lands heads.
- Estimate: The estimated probability 'p_hat' of heads is given by: p_hat = X/n

2. Confidence and Number of Flips:
The confidence interval, which provides a range where the true probability is likely to lie, can be given as:
    p_hat ± Z * sqrt(p_hat * (1 - p_hat) / n)
Where:
- p_hat is our estimate for the proportion of heads.
- Z is a factor based on desired confidence (e.g., for 95%, Z is approximately 1.96).

To target a specific margin of error 'E':
    E = Z * sqrt(p_hat * (1 - p_hat) / n)
If 'p_hat' is unknown, we can use p_hat = 0.5 to maximize variance.
''''

# Python Simulation:

import random
import matplotlib.pyplot as plt
import seaborn as sns

def flip_coin(p):
    #Simulate a coin flip, return 1 for heads, 0 for tails.
    return 1 if random.random() < p else 0

def simulate_flips(n, p_true):
    #Simulate 'n' coin flips and return the number of heads.
    return sum(flip_coin(p_true) for _ in range(n))

def experiment(n, p_true):
    #Estimate the probability of heads based on 'n' coin flips.
    return simulate_flips(n, p_true) / n

# Parameters:
n = 1000  # Number of coin flips for each experiment.
p_true = 0.7  # Actual probability of heads (unknown in real-world scenarios).

# Perform the simulation:
estimates = [experiment(n, p_true) for _ in range(10000)]

# Visualization:
sns.histplot(estimates, bins=50, kde=True, stat="density", alpha=0.5, color = 'green')
plt.axvline(p_true, color='red', linestyle='dashed', linewidth=1, label="True p")
plt.xlabel("Estimated Probability of Heads")
plt.ylabel("Density")
plt.title("Distribution of p Estimates from 10,000 Experiments")
plt.legend()
plt.show()
```

```python
import random
import math

def flip_coin(p):
    return 1 if random.random() < p else 0

# Parameters
p_true = 0.7  # true probability of heads

# Determine required sample size for 95% confidence and 5% margin of error
Z = 1.96
E = 0.05
p_hat_preliminary = 0.5
n_required = (Z**2 * p_hat_preliminary * (1 - p_hat_preliminary)) / E**2
n_required = int(math.ceil(n_required))  # Round up to ensure the margin of error requirement is met

# Conduct the experiment with n_required flips
heads_count = sum(flip_coin(p_true) for _ in range(n_required))
p_hat = heads_count / n_required

# Compute the 95% confidence interval
MoE = Z * math.sqrt(p_hat * (1 - p_hat) / n_required)
CI_lower = p_hat - MoE
CI_upper = p_hat + MoE

print(f"Required flips for 95% confidence and 5% margin of error: {n_required}")
print(f"Estimated p after {n_required} flips: {p_hat:.3f}")
print(f"95% Confidence Interval: ({CI_lower:.3f}, {CI_upper:.3f})")

```

```python
_='''
This code is designed to estimate the probability of a coin landing heads, when the true probability is unknown (our coin is bias). 
By performing the experiment 10,000 times with 'n' coin flips each iteration, we leverage the 
Law of Large Numbers which suggests that as the number of trials increases, the sample mean (or average outcome) 
approaches the expected (or true) value.

Furthermore, Our histogram's shape, with a large number of experiments, is influenced by the Central Limit Theorem (CLT). 
The CLT states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables 
approaches a normal (or Gaussian) distribution, regardless of the original distribution of the variables. This is why, even if our individual coin 
flips are binary and far from normally distributed, the averaged results over many flips start to form a bell curve.

The red dashed line in the visualization represents the true probability of the coin landing heads (set to 0.7 for this simulation). 
This serves as a benchmark to see how close our repeated experiments come to the true value.

One might question the choice of 10,000 experiments or the number 'n' for coin flips in each experiment. While larger numbers typically provide more 
accurate estimates, they also demand more computational resources. The chosen values represent a balance between accuracy and efficiency for this 
demonstration. However, in a real-world scenario, a more detailed analysis might be required to determine optimal values, especially considering factors 
like computational constraints and the desired confidence in the estimate.
'''
```
