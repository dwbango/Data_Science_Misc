```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the network as an undirected graph
G = nx.Graph()

# Add nodes (individuals in a social network)
nodes = ['Person A', 'Person B', 'Person C', 'Person D', 'Person E']
G.add_nodes_from(nodes)

# Add edges (interactions or connections between individuals)
edges = [('Person A', 'Person B'), ('Person A', 'Person C'), 
         ('Person B', 'Person C'), ('Person B', 'Person D'), 
         ('Person C', 'Person D'), ('Person C', 'Person E'), 
         ('Person D', 'Person E')]
G.add_edges_from(edges)

# Calculate Degree Centrality
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:")
print(degree_centrality)

# Calculate Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G, normalized=True)
print("\nBetweenness Centrality:")
print(betweenness_centrality)

# Calculate Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
print("\nCloseness Centrality:")
print(closeness_centrality)

# Calculate Prestige (using eigenvector centrality as a proxy)
prestige_centrality = nx.eigenvector_centrality(G)
print("\nPrestige (Eigenvector Centrality):")
print(prestige_centrality)

# Visualize the network
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # Seed for consistent layout

# Draw the network graph
nx.draw_networkx(
    G, pos, with_labels=True, 
    node_color='lightgreen', edge_color='gray', 
    node_size=2000, font_size=15
)

plt.title("Social Network Analysis (SNA) - Influence Network")
plt.axis("off")  # Turn off axis
plt.show()

```
