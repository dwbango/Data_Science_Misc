```python
Features = ['CNT_CHILDREN', 'AMT_INCOME_TOTAL',
       'DAYS_BIRTH', 'DAYS_EMPLOYED', 'FLAG_WORK_PHONE',
       'FLAG_PHONE', 'FLAG_EMAIL', 'CNT_FAM_MEMBERS',
       'CODE_GENDER_F', 'CODE_GENDER_M', 'FLAG_OWN_CAR_N', 'FLAG_OWN_CAR_Y',
       'FLAG_OWN_REALTY_N', 'FLAG_OWN_REALTY_Y',
       'NAME_INCOME_TYPE_Commercial associate', 'NAME_INCOME_TYPE_Pensioner',
       'NAME_INCOME_TYPE_State servant', 'NAME_INCOME_TYPE_Student',
       'NAME_INCOME_TYPE_Working', 'NAME_EDUCATION_TYPE_Academic degree',
       'NAME_EDUCATION_TYPE_Higher education',
       'NAME_EDUCATION_TYPE_Incomplete higher',
       'NAME_EDUCATION_TYPE_Lower secondary',
       'NAME_EDUCATION_TYPE_Secondary / secondary special',
       'NAME_FAMILY_STATUS_Civil marriage', 'NAME_FAMILY_STATUS_Married',
       'NAME_FAMILY_STATUS_Separated',
       'NAME_FAMILY_STATUS_Single / not married', 'NAME_FAMILY_STATUS_Widow',
       'NAME_HOUSING_TYPE_Co-op apartment',
       'NAME_HOUSING_TYPE_House / apartment',
       'NAME_HOUSING_TYPE_Municipal apartment',
       'NAME_HOUSING_TYPE_Office apartment',
       'NAME_HOUSING_TYPE_Rented apartment', 'NAME_HOUSING_TYPE_With parents']
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split

train, test = train_test_split(df7, test_size = 0.3, random_state=21)
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)

features_train = train[Features]
label_train = train['OverDues']
features_test = test[Features]
label_test = test['OverDues']
```

```python
#df7.drop('OCCUPATION_TYPE', axis=1, inplace=True)
```

```python
df7.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
import pandas as pd

# Initialize the Logistic Regression model
log_reg = LogisticRegression()

# Fit the model on the training data
log_reg.fit(features_train, label_train)

# Make predictions on the training and test sets
pred_train = log_reg.predict(features_train)
pred_test = log_reg.predict(features_test)

# Calculate and print accuracy for training and test sets
accuracy_train = accuracy_score(label_train, pred_train) * 100
accuracy_test = accuracy_score(label_test, pred_test) * 100
print('Training Accuracy is ' + str(round(accuracy_train, 4)) + '%')
print('Test Accuracy is ' + str(round(accuracy_test, 4)) + '%')

# Generate a confusion matrix
print('Confusion Matrix')
conf_matrix = confusion_matrix(label_test, pred_test)
columns = ['Non Defaulter', 'Defaulter']
print(pd.DataFrame(conf_matrix, columns=columns, index=columns), '\n')

# Print the classification report
print('Classification Report')
print(classification_report(label_test, pred_test), '\n')

# Additional metrics
print('Precision Score')
print(precision_score(label_test, pred_test, average=None), '\n')

print('Recall Score')
print(recall_score(label_test, pred_test, average=None), '\n')

print('F1 Score')
print(f1_score(label_test, pred_test, average=None), '\n')
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get predicted probabilities
y_pred_prob = log_reg.predict_proba(features_test)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(label_test, y_pred_prob)

# Calculate AUC
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace 'y_pred_prob' with the model's predicted probabilities for the positive class
# Replace 'label_test' with the actual binary class labels from the test set

# Sort the actual values by the predicted probabilities in descending order
sorted_indices = np.argsort(y_pred_prob)[::-1]
sorted_label_test = np.array(label_test)[sorted_indices]

# Calculate the cumulative sum of the positive class in the sorted labels array
cumulative_gains = np.cumsum(sorted_label_test)

# Normalize the cumulative gains by the total sum of positives to get a percentage
cumulative_gains = cumulative_gains / cumulative_gains[-1]

# Insert a zero at the beginning of the cumulative gains array for plotting
cumulative_gains = np.insert(cumulative_gains, 0, 0)

# Create an array with percentages of the total number of cases
cumulative_percentage = np.linspace(0, 1, len(cumulative_gains))

# Plot the cumulative gains chart
plt.figure(figsize=(10, 6))
plt.plot(cumulative_percentage, cumulative_gains, label='Model', color='blue')
plt.plot([0, 1], [0, 1], label='Baseline', linestyle='--', color='red')
plt.title('Cumulative Gains Chart')
plt.xlabel('Percentage of sample')
plt.ylabel('Cumulative gain')
plt.legend(loc='lower right')
plt.show()
```
