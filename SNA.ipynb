```python
import networkx as nx
import matplotlib.pyplot as plt

# Directed graph
G = nx.DiGraph()

edges = [
    ("A", "B"), ("A", "C"), ("B", "D"), 
    ("C", "D"), ("D", "E"), ("C", "E"), 
    ("E", "F"), ("F", "G"), ("G", "H"), 
    ("H", "I"), ("I", "J"), ("J", "A"), 
    ("B", "G"), ("D", "I"), ("H", "C")
]
G.add_edges_from(edges)

# Centrality measures
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G, normalized=True)
closeness_centrality = nx.closeness_centrality(G)
prestige = nx.in_degree_centrality(G)  # In-degree centrality as prestige

# Results
print("Degree Centrality:", degree_centrality)
print("Betweenness Centrality:", betweenness_centrality)
print("Closeness Centrality:", closeness_centrality)
print("Prestige (In-Degree Centrality):", prestige)

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Consistent layout for multiple runs

# Node sizes based on degree centrality
node_sizes = [v * 1200 for v in degree_centrality.values()]

# Node colors based on betweenness centrality
node_colors = [v for v in betweenness_centrality.values()]

# Nodes
nodes = nx.draw_networkx_nodes(G, pos, node_size=node_sizes, 
                               node_color=node_colors, cmap=plt.cm.viridis)

# Edges with directed arrows
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=15, edge_color='gray')

# Label
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

# Color bar for betweenness centrality
plt.colorbar(nodes, label="Betweenness Centrality")

# title and legend
plt.title('Advanced Directed Graph with Centrality Measures')

plt.show()
```
