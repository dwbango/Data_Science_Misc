```python
# Print your x (independent) and y (dependent) variables from the data frame. Show the results.
X = data2.drop(['SALE_PRC', 'log_SALE_PRC'], axis=1)  # Independent variables
y = data2['log_SALE_PRC']  # Dependent variable
print("Independent Variables (X):", X.columns)
print("Dependent Variable (Y):", y.name)
```

```python
# We tested for multicollinearity already
```

```python
# Split the data into 80% training and 20% testing sets using the train_test_split class.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
#Fit an Ordinary Least Squares (OLS) Model
import statsmodels.api as sm

X_train_const = sm.add_constant(X_train)  # Adds a constant term to the X matrix
ols_model = sm.OLS(y_train, X_train_const).fit()
print(ols_model.summary())
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Print the intercept and coefficients
print("Intercept:", lr_model.intercept_)
print("Coefficients:")
for var, coef in zip(X_train.columns, lr_model.coef_):
    print(f"{var}: {coef}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

#Make Predictions and Evaluate the Model

y_pred = lr_model.predict(X_test)

# Evaluation
print("R-squared:", r2_score(y_test, y_pred))
#print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
# Confusion Matrix

#Step 1: Categorize Predictions
# Define a threshold
threshold = 0.05 * y_test

# Categorize predictions as 1 if within 5% of the actual value, else 0
predictions_category = np.where(np.abs(y_pred - y_test) <= threshold, 1, 0)

# Categorize actual values as 1 (comparing the same values)
actual_category = np.ones_like(predictions_category)

#Step 2: Create a Simplified "Confusion Matrix"
TP = np.sum(predictions_category == 1)
FN = np.sum(predictions_category == 0)  # Not accurate based on threshold criterion

from sklearn.metrics import confusion_matrix

# Assuming predictions_category holds binary outcomes: 1 for accurate, 0 for not
conf_matrix = confusion_matrix(actual_category, predictions_category)
print(conf_matrix)
```

```python
# Accuracy Score
accuracy_score = TP / (TP + FN)
print("Conceptual 'Accuracy Score':", accuracy_score)
```

```python
# Validate the Model (k-fold cross validation)
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr_model, X, y, cv=5, scoring='r2')
print("Cross-validated R-squared scores:", scores)

scores_average = sum(scores) / len(scores)
print("Average of scores:", scores_average)
```
