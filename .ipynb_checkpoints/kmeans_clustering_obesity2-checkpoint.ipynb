```python
# Necessary Libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

To find the optimal number of clusters that best fits the dataset, I would use methods like the Elbow Method to identify the point where the decrease in the sum of squared distances within clusters becomes less pronounced, and the Silhouette Score to evaluate the separation and cohesion of the clusters formed at different values of k.

```python
# Selecting the features for clustering
features = obesity[['Height', 'Weight', 'Age']]
features = features.fillna(features.mean())  # Handling missing values
```

```python
# Standardizing the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Initialize k-means (I Chose 7 because of the 7 different weight classification levels in dataframe)
kmeans = KMeans(n_clusters=7, init='random', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(features_scaled)
```

```python
# Adding the cluster labels to dataframe
obesity['Cluster'] = clusters
```

```python
# Plot the clusters 
plt.scatter(obesity['Height'], obesity['Weight'], c=obesity['Cluster'], cmap='viridis')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Cluster of Obesity Data')
plt.show()
```

```python
# Printing centroids
print("Centroids:\n", kmeans.cluster_centers_)
```

```python
# calculate Euclidean distances
distances = euclidean_distances(features_scaled, kmeans.cluster_centers_)
```

```python
print(distances)
```

```python
# nearest centroid for the first data point
nearest_centroid_index = np.argmin(distances[0])
print("Nearest centroid index for the first data point:", nearest_centroid_index)
```

```python
# Assign clusters, update centroids, and check for convergence.
data_points = features_scaled 
centroids = np.array([[ 1.0449183, -0.30373012, -0.58270301],
                      [ 0.91501296, 1.40708069, -0.33329936],
                      [ 0.94188122, 0.78564984, 1.34102503],
                      [-0.03221409, -0.77450034, -0.61662575],
                      [-0.57561955, 0.48237618, 0.14805913],
                      [-0.82925481, -0.43204787, 2.23197878],
                      [-1.33125071, -1.12857729, -0.6070548]])

def k_means_clustering(data_points, centroids, max_iters=100):
    for iteration in range(max_iters):
        # Step 1: Assign clusters
        distances = euclidean_distances(data_points, centroids)
        clusters = np.argmin(distances, axis=1)

        # Step 2: Update centroids
        new_centroids = np.array([data_points[clusters == k].mean(axis=0) for k in range(centroids.shape[0])])

        # Check for convergence
        if np.allclose(centroids, new_centroids, atol=1e-4):
            break

        centroids = new_centroids

    return clusters, centroids

clusters, updated_centroids = k_means_clustering(data_points, centroids)

print("Updated Centroids:\n", updated_centroids)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(features_scaled, clusters)
print("Silhouette Score: ", score)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

distortions = []
K = range(1, 11)  
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(features_scaled)
    distortions.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Distortion (WCSS)')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
K = range(2, 20)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(features_scaled)
    score = silhouette_score(features_scaled, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(10, 6))
plt.plot(K, silhouette_scores, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for different numbers of clusters')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


k_values = [3, 7]
results = {}

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(features_scaled)
    
    # Calculate silhouette score
    silhouette_avg = silhouette_score(features_scaled, labels)
    
    # Store results
    results[k] = {
        'cluster_centers': kmeans.cluster_centers_,
        'labels': labels,
        'inertia': kmeans.inertia_,  # This is the WCSS
        'silhouette_score': silhouette_avg
    }

    print(f"Results for k={k}:")
    print(f"WCSS (Inertia): {results[k]['inertia']}")
    print(f"Silhouette Score: {results[k]['silhouette_score']}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Range of k values
K = range(2, 21)

# Lists to store metrics
wcss = []
silhouette_scores = []

# Compute metrics for each k
for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(features_scaled)
    wcss.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(features_scaled, labels)
    silhouette_scores.append(silhouette_avg)
    print(f'k={k}: WCSS={kmeans.inertia_:.2f}, Silhouette Score={silhouette_avg:.4f}')

# Plot WCSS (Elbow Method)
plt.figure(figsize=(10, 5))
plt.plot(K, wcss, 'bo-')
plt.xlabel('Number of Clusters, k')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method: WCSS vs. Number of Clusters')
plt.grid(True)
plt.show()

# Plot Silhouette Scores
plt.figure(figsize=(10, 5))
plt.plot(K, silhouette_scores, 'ro-')
plt.xlabel('Number of Clusters, k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores vs. Number of Clusters')
plt.grid(True)
plt.show()
```

# reasoning for maybe opting for Simplicity and Clarity with k=3 (hypothetically)

In the analysis of obesity levels through k-means clustering, choosing k=3 presents a strategic approach to segmenting the dataset into three distinct weight classifications: underweight, normal weight, and overweight/obese. This choice is guided by the principles of interpretability and practical applicability in health-related contexts.

Simplified Interpretation: Clustering into three broad categories aligns well with common medical practices that categorize weight status into fundamental groups. This simplification facilitates clearer communication and understanding among healthcare providers and patients, enabling straightforward interpretations and recommendations based on easily identifiable groups.

Avoiding Overfitting: Higher values of k, such as 15, although statistically justifiable in terms of silhouette scores, risk overfitting the model to the data. Overfitting in clustering could lead to creating overly specific groups that may not have practical significance or reproducibility in other datasets or real-world scenarios. It may capture random variations or noise in the dataset as meaningful distinctions, which can mislead intervention strategies.

Reducing Complexity: A lower number of clusters reduces the complexity of the model, making it more robust and easier to manage. It avoids the granular fragmentation of the population into too many groups, which can complicate the development and implementation of public health policies and interventions. With k=3, the model focuses on major distinctions across the weight spectrum, supporting broad, impactful health initiatives rather than nuanced, difficult-to-implement actions.

Practical Utility: By focusing on three primary categories, interventions can be more generalized yet effective, targeting significant and common issues related to each weight classification. This approach ensures that resources are optimally allocated, enhancing the overall efficiency of health programs.
