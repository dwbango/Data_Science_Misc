```python
from nba_api.stats.static import players

# Get all players.
all_players = players.get_players()

# Find player by name, example: LeBron James
lebron = [player for player in all_players if player['full_name'] == 'LeBron James'][0]
print(lebron['id'])  # Prints LeBron's player_id
```

```python
from nba_api.stats.static import teams

# Get all teams.
all_teams = teams.get_teams()

# Find a team by name, example: Los Angeles Lakers
lakers = [team for team in all_teams if team['full_name'] == 'Los Angeles Lakers'][0]
print(lakers['id'])  # Prints the ID of the Lakers
```

```python
from nba_api.stats.endpoints import teamgamelog

# Fetch game logs for the Lakers
lakers_games = teamgamelog.TeamGameLog(team_id=lakers['id'])
print(lakers_games.get_data_frames()[0])  # Print the game logs as DataFrame
```
