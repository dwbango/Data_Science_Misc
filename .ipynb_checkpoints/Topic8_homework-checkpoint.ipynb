```python
#Assumption: No influential outliers
import numpy as np

def detect_outliers_iqr(data):
    outliers = []
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    outlier_step = 1.5 * IQR

    lower_bound = Q1 - outlier_step
    upper_bound = Q3 + outlier_step

    # Generate a list of indices of outliers
    outliers = data.index[(data < lower_bound) | (data > upper_bound)]

    return outliers

# Apply the function to each feature column in the df to check for outliers
for feature in data.columns:
    if data[feature].dtype != 'O':  # Skip non-numeric columns (excludes columns with data type 'O', a.k.a. "object.")
        outlier_indices = detect_outliers_iqr(data[feature])
        print(f"Outliers for {feature} are at index: {outlier_indices}")
```

```python
def remove_outliers_iqr(data):
    outlier_indices_list = []
    
    for feature in data.columns:
        if data[feature].dtype != 'O':  # Skip non-numeric columns
            Q1 = np.percentile(data[feature], 25)
            Q3 = np.percentile(data[feature], 75)
            IQR = Q3 - Q1
            outlier_step = 1.5 * IQR
            
            lower_bound = Q1 - outlier_step
            upper_bound = Q3 + outlier_step
            
            # Find the indices of outliers for the feature and append them to a list
            feature_outliers = data.index[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
            outlier_indices_list.extend(feature_outliers)
    
    # Select outliers indices, make sure the frequency is above a threshold
    outlier_indices_list = list(set(outlier_indices_list))  # Removes duplicates
    
    # Drop outliers
    data_cleaned = data.drop(outlier_indices_list)
    
    return data_cleaned

# apply the function to dataframe
data_cleaned = remove_outliers_iqr(data)

# shape of the data before and after removal
print(f"Original Data: {data.shape}")
print(f"Data after Outlier Removal: {data_cleaned.shape}")

# recheck any violated assumptions and rerun the model with the cleaned dataset
```

```python
'''
1. Introduction

This report details the statistical analysis conducted on a dataset comprising diamond attributes and their 
corresponding prices. The aim is to develop a regression model to predict diamond prices based on several predictors
such as carat, cut, color, clarity, depth, table, x (length), y (width), and z (depth). Understanding these relationships
aids stakeholders in pricing diamonds more accurately.

2. Data Preprocessing

The data was rigorously cleaned to ensure quality analysis. Steps taken included checking for missing values, 
which were not present in this dataset, and removing outliers using the Interquartile Range (IQR) method to mitigate
their impact on the model's performance. The resulting dataset contained 47,527 observations, reduced from the 
original 53,943.

3. Exploratory Data Analysis (EDA)

Summary statistics were computed to understand data distribution, and visual plots were created to observe relationships 
between features. This initial analysis revealed a strong positive correlation between carat and price, and between the 
dimensions (x, y, z) and carat.

4. Correlation Analysis

The correlation matrix indicated a potential multicollinearity issue, especially between the carat and the dimensions of the 
diamonds. A Variance Inflation Factor (VIF) analysis further confirmed multicollinearity concerns. Hence, the dimensions 
(x, y, z) were removed from the model to reduce multicollinearity.

5. Regression Modeling

An Ordinary Least Squares (OLS) regression model was selected for its interpretability and common use in predictive
pricing models. A reduced model was built using 'carat', 'depth', and 'table' as independent variables, which are 
significant in predicting diamond prices.

6. Model Diagnostics

Diagnostic checks showed that the residuals were independently and normally distributed. The assumptions of linearity and 
homoscedasticity were also met, confirming that the OLS model was appropriate for this data.

7. Results

The OLS regression results indicated a high R-squared value of 0.854, showing that the model explains a significant portion 
of the variance in diamond prices. The 'carat' variable had the highest positive coefficient, underscoring its importance in 
predicting the price. Conversely, 'depth' and 'table' had smaller and negative coefficients.

8. Limitations and Recommendations

The analysis has limitations, such as the exclusion of potentially influential categorical variables 
(e.g., cut, color, clarity) due to focusing on numerical predictors. Future analysis could include these variables as 
part of a more comprehensive model. I would like to conduct ANOVA to compare the mean prices between the clarity and 
color categorical variables.

9. Conclusion

The regression model showed that carat weight is the most significant predictor of diamond prices, while depth and table
have lesser yet significant roles. The study confirms that while size matters greatly in valuing diamonds, other factors 
like depth and table also play roles in pricing.
'''
```
