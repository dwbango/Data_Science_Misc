```python
# Outlier Check

# Calculate Z-scores
z_scores = stats.zscore(data)

# Find outliers
outliers_z = (abs(z_scores) > 3)

# Z-score outliers
#print("Outliers based on Z-scores:")
#print(data[outliers_z.any(axis=1)])

# Calculate IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Find outliers
outliers_iqr = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))

# IQR outliers
#print("Outliers based on IQR:")
#print(data[outliers_iqr.any(axis=1)])
```

```python
#Outliers Check

# Calculate Z-scores for each column
z_scores = stats.zscore(data[['sgot', 'sgpt']])

# Convert to DataFrame
z_scores_df = pd.DataFrame(z_scores, columns=['sgot', 'sgpt'], index=data.index)

# Find outliers in each column
outliers_sgot = z_scores_df['sgot'].abs() > 3
outliers_sgpt = z_scores_df['sgpt'].abs() > 3

# Filter outliers for 'sgot' and 'sgpt'
outliers_sgot_rows = data[outliers_sgot]
outliers_sgpt_rows = data[outliers_sgpt]

# Print outliers for 'sgot'
print("Outliers in SGOT:")
print(outliers_sgot_rows)

print("\n---------------------------\n")

# Print outliers for 'sgpt'
print("Outliers in SGPT:")
print(outliers_sgpt_rows)
```

```python
#Removing outliers based on z-score method
# Combine the outlier for 'sgot' and 'sgpt'
outliers_combined = outliers_sgot | outliers_sgpt

# Invert the condition to keep non-outlier rows
non_outliers = ~outliers_combined

# Filter the data to remove outliers
data_no_outliers = data[non_outliers]

# Print the dataframe without outliers
print(data_no_outliers)

num_rows1 = len(data_no_outliers) # Should be 9 less observations
print("Number of rows:", num_rows1)
```

```python
# Simple Linear Regression:
import statsmodels.api as sm

X = data['sgpt']  # Independent variable
y = data['sgot']  # Dependent variable

# Adding a constant to the model for the intercept
X = sm.add_constant(X)

# Build and fit the model
model = sm.OLS(y, X).fit()

# Print the summary of the regression (display results)
print(model.summary())
```

```python
# Scatter plot with Regression
# Linearity Check
plt.figure(figsize=(10, 5))
sns.scatterplot(x='sgpt', y='sgot', data=data, color='blue')

# Plotting the regression line
plt.plot(X['sgpt'], model.predict(X), color='red')

# Labels
plt.title('SGOT vs SGPT')
plt.xlabel('SGPT')
plt.ylabel('SGOT')
plt.show()
```

```python
# Residuals vs Fitted Values for Homoscedasticity
plt.figure(figsize=(10, 6))
sns.residplot(x=model.predict(X), y=model.resid, lowess=True, line_kws={'color': 'red', 'lw': 1})
plt.title('Residuals vs Fitted')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.show()
```

```python
# Q-Q plot for normality of residuals
sm.qqplot(model.resid, line ='45', fit=True)
plt.title('Normal Q-Q')
plt.show()

# density plot for the residuals
sns.kdeplot(model.resid, fill=True)
plt.title('Density Plot of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Density')
plt.show()

# Shapiro-Wilk test for normality
shapiro_test = stats.shapiro(model.resid)

print(f"Shapiro-Wilk Test: Statistic={shapiro_test[0]}, p-value={shapiro_test[1]}")

#Shapiro statistic value close to 1 and an p-value grater than .05 would indicate
#normally distributed residuals: This is not the case for our data.
```

```python
#Independence of residuals
from statsmodels.stats.stattools import durbin_watson

# Calculate Durbin-Watson statistic
dw_statistic = durbin_watson(model.resid)

# Print the result
print(f'Durbin-Watson Statistic: {dw_statistic}')

#Durbin-Watson statistic is close to 2, 
#which suggests that there is little to no autocorrelation in the residuals of the model 
```

```python
# Transforming the data to meet normality assumption:

# Apply Box-Cox transformation to 'sgot'
data['sgot_transformed'], lmbda_sgot = stats.boxcox(data['sgot'])

# Apply Box-Cox transformation to 'sgpt'
data['sgpt_transformed'], lmbda_sgpt = stats.boxcox(data['sgpt'])

# Print the lambda values
print(f"Lambda for SGOT transformation: {lmbda_sgot}")
print(f"Lambda for SGPT transformation: {lmbda_sgpt}")

# Create model then assess the normality assumption and others with transformed data
```
