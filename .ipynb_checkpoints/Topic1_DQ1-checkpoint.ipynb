```python
import pandas as pd
pd.set_option('display.max_columns', None)

import matplotlib.pyplot as plt
```

```python
# Statistical Concepts Explanation:

# Population:
# The entire group or set that is the subject of the study, encompassing all members or data points of interest.
#Example: All QB's that have played during the 2023-2024 NFL season

# Sample:
# A subset of the population chosen for the study. 
#It represents a smaller group selected to infer about the entire population.
#Example: All QBs that play in the AFC 

# Parameter:
# A numerical value or characteristic describing a feature of the entire population.
# Example: The AVG QBR of all current NFL QBs would be a population parameter.

# Statistic:
# A numerical value or characteristic from a sample, often used to estimate a population parameter.
# Example: AVG QBR of QBs currently playing in the AFC is a sample statistic.

#Note: 
#QBR is a metric.
#The average QBR across all NFL QBs is a population parameter.
#The average QBR for just AFC QBs is a sample statistic.
```

```python
# Team Management Decisions:
# Scenario: Management aims to increase the team's win rate, identifying the QB as a pivotal factor.
# Use of Statistics: Analyzing league-wide statistics to pinpoint metrics (like QBR) that correlate with team wins.
# Actionable Decision: During free agency or trades, aim for QBs with high QBR to increase chances of more victories.

# Sports Betting:
# Scenario: Bettors aim to optimize their betting odds using data-driven insights.
# Use of Statistics: Analyze current season statistics to determine teams or QBs that perform above average.
# Actionable Decision: Place bets on teams with high QBR QBs, especially against teams with weaker defenses.

# Fan Engagement & Marketing:
# Scenario: Marketing aims to enhance team merchandise sales and boost fan engagement.
# Use of Statistics: Recognize trends where high-QBR QBs correlate with better merchandise sales and fan engagement.
# Actionable Decision: Increase marketing campaigns around star QBs, launch QB-focused merchandise, and roll out 
# targeted advertising campaigns.

# Sports Analytics (Focusing on NFL QBs)
# Data Filtering & Comparison: With pandas in Python, you can filter data to focus on particular teams, players, or timeframes. 
# It is also useful to compare one QB's stats against another or against an average.

# Visualization: Libraries like matplotlib enable you to visualize the data in intuitive ways.
#tools like histograms can be used to visualize the distribution of a single variable 
# or Scatter plots can show relationships between two variables like yards and TDs 

# Statistical Analysis with Python:
# With Python, we can conduct t-tests, ANOVAs, and build predictive models 
# to determine the factors that influence key outcomes. 
# This allows decisions to be based on statistical evidence rather than merely on observable trends.
```

```python
df1 = pd.read_csv("downloads/passing23.csv")
df1 = df1[df1['Att'] >= 10]
df1.fillna(0, inplace=True)
#df1.head()
#df1
```

```python
#Creating samples
# Define a list of AFC & NFC teams by their abbreviations
afc_teams = ["BAL", "BUF", "CIN", "CLE", "DEN", "HOU", "IND", "JAX", "KAN", "LAC", "LVR", "MIA", "NWE", "NYJ", "PIT", "TEN"]
nfc_teams = ["DAL", "PHI", "WAS", "NYG", "GNB", "MIN", "CHI", "DET", "NOR", "TAM", "CAR", "ATL", "SEA", "SFO", "LAR", "ARI"]



# Filter df1 to get only the rows (QBs) that belong to AFC teams
afc_qbs = df1[df1['Tm'].isin(afc_teams)]

# Filter df1 to get only the rows (QBs) that belong to NFC teams
nfc_qbs = df1[df1['Tm'].isin(nfc_teams)]

# Display the AFC QBs sample in a table
#afc_qbs

# Display the first few rows of the NFC QBs sample
#print(nfc_qbs.head())

#make sure all teams are accounted for and abbreviations align
unique_teams_count = nfc_qbs['Tm'].nunique()
print(unique_teams_count)
unique_teams_count1 = afc_qbs['Tm'].nunique()
print(unique_teams_count1)
```

```python
#population parameter
avg_qbr_all_qbs = df1['QBR'].mean()
print(f"The average QBR for all QBs this season is {avg_qbr_all_qbs:.2f}")
```

```python
#Sample Statistics
# Average QBR for AFC QBs
avg_qbr_afc_qbs = afc_qbs['QBR'].mean()
print(f"The average QBR for AFC QBs this season is {avg_qbr_afc_qbs:.2f}")

# Average QBR for NFC QBs
avg_qbr_nfc_qbs = nfc_qbs['QBR'].mean()
print(f"The average QBR for NFC QBs this season is {avg_qbr_nfc_qbs:.2f}")
```

```python
#Visual Aid
plt.scatter(afc_qbs['Yds'], afc_qbs['TD'], color='blue', alpha=0.5)
plt.title('Yards vs. Touchdowns for AFC QBs')
plt.xlabel('Yards')
plt.ylabel('Touchdowns')
plt.show()
```

```python
#This scatter plot represents the relationship between two sample statistics from the sample of AFC QBs.
#The x-axis represents the total passing yards (a sample statistic) for each AFC QB.
#The y-axis represents the total number of touchdowns (another sample statistic) for each AFC QB.
```

```python
# Scatter plot with custom markers
plt.scatter(afc_qbs['Yds'], afc_qbs['TD'], 
            s=25,  # set marker size to 50
            marker='D',  # set marker shape to diamond
            edgecolors='red',  # red edge color
            facecolors='blue',  # blue fill color
            alpha=0.5)  # semi-transparent

plt.title('Yards vs. Touchdowns for AFC QBs')
plt.xlabel('Yards')
plt.ylabel('Touchdowns')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 4)) # set the size that you'd like (width, height)

ax.axis('off')

# The table takes a 2D array as input:
table_data = []
table_data.append(afc_qbs.columns)
table_data.extend(afc_qbs.values.tolist())

table = ax.table(cellText=table_data, loc='center', cellLoc = 'center', colWidths=[0.1]*afc_qbs.shape[1])
table.auto_set_font_size(False)
table.set_fontsize(10)
plt.show()
```
