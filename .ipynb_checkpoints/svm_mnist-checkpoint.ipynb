# Applying SVM Classification Method

By applying a SVM classification method to the MNIST dataset, we aim to:
- Evaluate the algorithm's accuracy in differentiating between handwritten digits.
- Identify which digits or features contribute most to misclassifications.
- Determine the model's strengths and weaknesses across variations in digit presentations.

***SVM's Uniqueness and Application***

The Support Vector Machine (SVM) algorithm stands out due to its unique method of categorizing data by hyperplanes, aiming to maximize the margin between different classes. This strategy is particularly effective for high-dimensional data classification, enabling SVM to handle complex datasets. The process involves:
- Data preprocessing, including normalization and reshaping.
- Utilizing a linear kernel to project data into a higher-dimensional space, simplifying non-linear classification problems.
- Training the SVM to identify the optimal hyperplane that achieves the maximum separation between classes.
- Evaluating model performance based on accuracy and misclassifications.

***Kernel, Hyperplane, and Decision Boundary in SVM***

- **Kernel**: A function that enables SVMs to perform in high-dimensional spaces without the explicit transformation of data, allowing for efficient handling of non-linear separations.
- **Hyperplane**: A decision plane, the hyperplane is the main component of SVM's classification capability, dividing data points into distinct classes in the feature space.
- **Decision Boundary**: Formed by the hyperplane, this boundary identifies where the classification decision decyphers from one class to another.

***Maximum Margin, Support Vectors, and Maximum Margin Hyperplane***

- **Maximum Margin:** The widest possible margin between data points of different classes, aiming for the greatest separation to reduce misclassification.
- **Support Vectors:** Data points closest to the hyperplane, which influence its position and orientation. Changing these points would change the hyperplane.
- **Maximum Margin Hyperplane:** The optimal hyperplane that separates classes with the maximum margin, central to SVM's classification strategy for robustness against future data.

***Roles in SVM classification***:
Each component plays an important role in defining SVM's classification logic, with the maximum margin ensuring minimal classification error, support vectors deciding the hyperplane's placement, and the maximum margin hyperplane serving as the decision-maker in class separation.

# Required Libraries

```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
#!pip install tensorflow
#from tensorflow.keras.datasets import mnist
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from itertools import cycle
from sklearn.model_selection import cross_val_score
```

# Loading the MNIST Dataset

```python
# Import necessary libraries
#from tensorflow.keras.datasets import mnist

# Load dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Select a subset for training
random_indices_train = np.random.choice(train_images.shape[0], 8000, replace=False)
x_train_sub = train_images[random_indices_train]
y_train_sub = train_labels[random_indices_train]

# Select a subset for testing
random_indices_test = np.random.choice(test_images.shape[0], 2000, replace=False)
x_test_sub = test_images[random_indices_test]
y_test_sub = test_labels[random_indices_test]

print("Subset shapes:")
print("Training data: ", x_train_sub.shape, "Training labels: ", y_train_sub.shape)
print("Testing data: ", x_test_sub.shape, "Testing labels: ", y_test_sub.shape)
```

The MNIST dataset subset for this project was created and consists of 10,000 images. 8,000 for training and 2,000 for testing, all images are 28x28 pixels. This subset was randomly selected from the original training and test sets for manageability with my computing power. With 80% of the images used for training and 20% for testing, this split is designed to effectively develop and evaluate the model.

```python
# Check if there are any NaN values in the data
print(np.isnan(x_train_sub).any())
print(np.isnan(y_train_sub).any())
print(np.isnan(x_test_sub).any())
print(np.isnan(y_test_sub).any())
```

```python
# Flatten the images
x_train_flat = x_train_sub.reshape((x_train_sub.shape[0], -1))
x_test_flat = x_test_sub.reshape((x_test_sub.shape[0], -1))
```

```python
# Normalize the pixel values
x_train_norm = x_train_flat / 255.0
x_test_norm = x_test_flat / 255.0
```

In the MNIST dataset, all images are pre-processed to a uniform size of 28x28 pixels, ensuring consistency across the dataset. Each pixel's value ranges from 0 to 255, representing the grayscale intensity. Before analysis, we normalized these pixel values by dividing by 255, converting them to a scale of 0 to 1. This step is crucial for machine learning models as it ensures that the gradient descent algorithm converges more efficiently during training by dealing with smaller and more manageable values. Our exploration confirmed that there are no missing values, outliers or unusual phenomena within the dataset: all digit labels correctly fall within the expected range of 0-9, and the distribution of these labels is relatively uniform, indicating a balanced dataset. This balance is vital for training our SVM classifier, as it prevents bias towards a more frequently represented classes.

# Training the SVM Model on MNIST Data

```python
# Define the SVM classifier (linear kernel)
classifier = svm.SVC(kernel = 'linear')
```

```python
# Train the classifier on the training set
classifier.fit(x_train_norm, y_train_sub) 
```

# SVM Classifier Evaluation

```python
# Make predictions on the test set
y_pred = classifier.predict(x_test_norm)
```

```python
# Calculate accuracy on the test set
accuracy = accuracy_score(y_test_sub, y_pred)
print("Test Accuracy:", accuracy)
```

```python
# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test_sub, y_pred))
```

```python
# Classification Report
print("\nClassification Report:")
print(classification_report(y_test_sub, y_pred))
```

```python
# ROC-AUC Curves
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from itertools import cycle

# Binarize the output labels for ROC computation
y = label_binarize(y_test_sub, classes=[i for i in range(10)])
n_classes = y.shape[1]

# Learn to predict each class against the other
classifier = OneVsRestClassifier(SVC(kernel='linear', probability=True))
y_score = classifier.fit(x_train_norm, y_train_sub).predict_proba(x_test_norm)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves
colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'red', 'green', 'blue', 'black', 'pink', 'brown', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve of class {0} (area = {1:0.2f})'.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-Class ROC Curves for SVM Classification of MNIST Dataset')
plt.legend(loc="lower right")
plt.show()
```

```python
# k-fold cross validation

# Define the SVM classifier with a linear kernel
classifier = svm.SVC(kernel='linear')

# Perform 5-fold cross-validation and calculate the mean accuracy
scores = cross_val_score(classifier, x_train_norm, y_train_sub, cv=5)

# Print the accuracy for each fold and the mean accuracy
print("Accuracy for each fold: ", scores)
print("Mean cross-validation accuracy: ", scores.mean())
```
