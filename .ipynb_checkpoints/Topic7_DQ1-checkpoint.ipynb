```python
_'''
Both linear regression and ANOVA are statistical methods that help in understanding relationships between
variables and can be used to test hypotheses about these relationships. Typically, you will use a linear 
regression model when you aim to predict the value of a dependent variable that is continuous. Although having 
independent quantitative variables is easier for this method, linear regression accommodates both quantitative 
and categorical predictor (independent) variables. When you have categorical predictor variables, encoding 
techniques like dummy or effect coding are necessary. On the other hand, you will use the ANOVA test when you 
aim to compare the means of a dependent variable across different levels or categories of an independent variable. 
It is used to determine if there exists a statistically significant difference in the means of three or more groups. 
While the dependent variable should be continuous, the independent variable in ANOVA is categorical.

In short, Choose ANOVA to compare means across multiple groups of a categorical independent variable, and 
choose linear regression to predict a continuous dependent variable based on one or more predictors 
(quantitative or categorical).

Both ANOVA and linear regression assume homoscedasticity (equal variances), independence of observations, 
and a normally distributed dependent variable (especially for smaller sample sizes). 
However, linear regression additionally assumes a linear relationship between predictors and the dependent variable.
'''

```

```python
import pandas as pd

# Load in the Dataset and cleaning
data = pd.read_csv('/Users/don/Downloads/Student_Performance.csv')

```

```python
print(data.head())
#print(data.tail())
print(data.columns)
```

```python
import statsmodels.api as sm

# Extracting predictor and target variables
X = data[['Hours Studied', 'Previous Scores', 'Sleep Hours', 'Sample Question Papers Practiced']]
X = sm.add_constant(X)  # Adding a constant term to the predictor variables
y = data['Performance Index']

# Fitting the model
model = sm.OLS(y, X).fit()

# Printing the summary
print(model.summary())
```

```python
# No multicollinearity: The independent variables should not be too highly correlated with each other.
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Drop the constant
X_vif = X.drop('const', axis=1)

# Calculate VIF for each predictor variable
vif_data = pd.DataFrame()
vif_data['Feature'] = X_vif.columns
vif_data['VIF'] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print(vif_data)
```
