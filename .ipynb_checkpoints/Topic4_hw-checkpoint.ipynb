```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
import matplotlib.ticker as mticker
```

```python
standings = pd.read_csv("../downloads/NFL.csv")
print(standings.head())
#print(standings)
```

```python
scoring_offense = pd.read_csv("../downloads/scoring.csv")
scoring_offense = pd.merge(scoring_offense, standings[['Tm', 'W']], on='Tm', how='inner')
print(scoring_offense.columns)
```

```python
conversions = pd.read_csv("../downloads/conversions.csv", skiprows=1)
# Add wins column
conversions = pd.merge(conversions, scoring_offense[['Tm', 'AllTD']], on='Tm', how='inner')
# Add AllTD column
conversions = pd.merge(conversions, standings[['Tm', 'W']], on='Tm', how='inner')
# Remove the 'RK' column
conversions = conversions.drop('Rk', axis=1)
# Replace '%' with an empty string for the entire dataframe and convert applicable columns to float type
conversions = conversions.replace({'%': ''}, regex=True)
conversions[['3D%', '4D%', 'RZPct']] = conversions[['3D%', '4D%', 'RZPct']].astype(float)  

print(conversions.columns)
print(conversions)
```

```python
# Tm / Games Played 

plt.figure(figsize=(14, 8))

# Bar chart for teams vs wins
sns.barplot(data=conversions, x='Tm', y='W', palette='Set2', errorbar=None)

# Title and labels
plt.title('Teams vs Wins')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)  # Rotating team names for better visibility

plt.tight_layout()
plt.show()
```

```python
# Tm / Games Played 
plt.figure(figsize=(14,6))
sns.barplot(x='Tm', y='G', data=conversions, palette='viridis')
plt.title('Number of Games Played by Each Team')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
plt.show()
```

```python
#3DAtt
plt.figure(figsize=(10,6))
sns.histplot(conversions['3DAtt'], kde=True, bins=10, color='red')
plt.title('Distribution of 3rd Down Attempts across Teams')
plt.xlabel('Number of 3rd Down Attempts')
plt.ylabel('Number of Teams')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python

#3DAtt
plt.figure(figsize=(15,8))
sorted_teams = conversions.sort_values(by='3DAtt', ascending=False)
sns.barplot(x=sorted_teams['Tm'], y=sorted_teams['3DAtt'], palette='viridis')
plt.title('3rd Down Attempts by Team')
plt.xlabel('Team')
plt.ylabel('Number of 3rd Down Attempts')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
#3DConv
plt.figure(figsize=(12, 7))
sns.scatterplot(x=conversions['3DConv'], y=conversions['AllTD'], hue=conversions['Tm'], palette='Set2', s=100)
plt.title('3rd Down Conversions vs. AllTD by Team')
plt.xlabel('3rd Down Conversions')
plt.ylabel('AllTD')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
#3DConv
plt.figure(figsize=(15,8))
sorted_teams = conversions.sort_values(by='3DConv', ascending=False)
sns.pointplot(x=sorted_teams['Tm'], y=sorted_teams['3DConv'], palette='magma', linestyles="--", markers="o")
plt.title('3rd Down Conversions by Team')
plt.xlabel('Team')
plt.ylabel('Number of 3rd Down Conversions')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
#3D%
# Calculate the correlation
corr_val = conversions['3D%'].corr(conversions['W'])

plt.figure(figsize=(12, 7))

# Scatter plot
sns.scatterplot(data=conversions, x='3D%', y='W', hue='Tm', palette='Set2', s=100)

# Horizontal line for average 3rd Down Percentage
avg_3d_pct = conversions['3D%'].mean()
plt.axvline(x=avg_3d_pct, color='red', linestyle='--', label=f'Average 3D% = {avg_3d_pct:.1f}%')  # Changed format here

plt.title(f'3rd Down Conversion Percentage vs. Wins by Team\nCorrelation: {corr_val:.2f}')
plt.xlabel('3rd Down Conversion Percentage')
plt.ylabel('Wins')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

```python
#3D%
plt.figure(figsize=(14, 8))

# Sort teams by wins and select top 3 and bottom 3
top_teams = conversions.sort_values(by='W', ascending=False).head(3)
bottom_teams = conversions.sort_values(by='W', ascending=True).head(3)

# Calculate average 3D conversion percentages
top_avg_3d_pct = top_teams['3D%'].mean()
bottom_avg_3d_pct = bottom_teams['3D%'].mean()
league_avg_3d_pct = conversions['3D%'].mean()  # Calculating league average

# Combine top and bottom teams
combined_teams = pd.concat([top_teams, bottom_teams])

# Plot
sns.barplot(data=combined_teams, x='Tm', y='3D%', palette='viridis', order=combined_teams['Tm'])

# Display averages with horizontal lines
plt.axhline(y=top_avg_3d_pct, color='blue', linestyle='--', label=f'Average 3D% of Top Teams = {top_avg_3d_pct:.1f}%')
plt.axhline(y=bottom_avg_3d_pct, color='red', linestyle='--', label=f'Average 3D% of Bottom Teams = {bottom_avg_3d_pct:.1f}%')
plt.axhline(y=league_avg_3d_pct, color='green', linestyle='-', label=f'League Average 3D% = {league_avg_3d_pct:.1f}%')  # Displaying league average

# Title and labels
plt.title('3D Conversion Percentage of Top 3 and Bottom 3 Teams by Wins')
plt.xlabel('Team')
plt.ylabel('3D Conversion Percentage')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

```

```python
#4DAtt


plt.figure(figsize=(14, 8))

# Group by wins and calculate the mean of 4DAtt for each group
grouped_data = conversions.groupby('W')['4DAtt'].mean().reset_index()

# Plotting
sns.lineplot(data=grouped_data, x='W', y='4DAtt', marker="o")

# Title and labels
plt.title('Average 4th Down Attempts by Number of Wins')
plt.xlabel('Wins')
plt.ylabel('Average 4th Down Attempts')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(grouped_data['W'])  #  each win value is a tick on the x-axis

# y-axis values have two decimal places
y_format = mticker.FuncFormatter(lambda x, _: '{:.2f}'.format(x))
plt.gca().yaxis.set_major_formatter(y_format)

plt.tight_layout()
plt.show()

```

```python
# RZTD
# Sort the dataframe by RZTD and extract top 8 teams
top_8_teams = sorted_conversions.head(8)
teams = top_8_teams['Tm'].tolist()
rztd_values = top_8_teams['RZTD'].tolist()

# Calculate the sum of RZTD for the rest of the teams and append it to our data
rest_sum = sorted_conversions['RZTD'].iloc[8:].sum()
teams.append('Other Teams')
rztd_values.append(rest_sum)

# Create the pie chart for the top 8 teams and the rest grouped together
plt.figure(figsize=(12, 7))
colors = sns.color_palette('Set3', len(teams))
plt.pie(rztd_values, labels=teams, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Share of Red Zone Touchdowns by Top 8 Teams and Other Teams Combined')
plt.show()

# Sum of RZTD values for the top 8 teams
top_8_sum = top_8_teams['RZTD'].sum()

# Total RZTD values in the dataset
total_rztd = conversions['RZTD'].sum()

# Calculate the percentage
top_8_percentage = (top_8_sum / total_rztd) * 100

print(f"The top 8 teams account for {top_8_percentage:.2f}% of the total Red Zone Touchdowns.")
```
