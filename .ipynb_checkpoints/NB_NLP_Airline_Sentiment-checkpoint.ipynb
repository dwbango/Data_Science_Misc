# The Naive Bayes Algorithm and Its Application

The Naive Bayes classifier is a probabilistic machine learning model used for sentiment analysis, valued for its simplicity and efficiency in dealing with large datasets of textual data.

**Principle**: Naive Bayes applies Bayes’ theorem under the assumption that each feature (word) is conditionally independent of every other feature. While this independence assumption is rarely true in natural language, Naive Bayes often performs remarkably well for text classification because this simplified model still captures enough relevant structure for accurate predictions.

**Model Selection**: Given the textual nature of our dataset, we plan to employ the **Multinomial Naive Bayes** variant. This choice is driven by the algorithm’s suitability for discrete features (such as word counts or frequencies) produced after vectorizing tweets. Multinomial Naive Bayes is particularly adept at handling frequency data, aligning well with the requirements of this classification task.

**Key Considerations for Naive Bayes**:
- **Assumption of Independence**: Although not strictly valid for natural language (where words can be highly correlated), the model’s performance in text classification remains surprisingly robust.
- **Data Preprocessing**: Effective text preprocessing—such as tokenization, removing stop words, and optionally stemming or lemmatization—is essential to reduce dimensionality and focus on the most informative aspects of the text.

By employing the Naive Bayes classifier, the aim is to categorize tweets into positive, negative, or neutral sentiments accurately. This analysis will help us gauge public sentiment toward various airlines, providing insights into both existing strengths and opportunities for improvement in customer feedback.

```python
# Cleaning tweet text by removing special characters for readability

# Function to Clean Tweet Text
def clean_text(text):
    '''Cleans the input text by applying several regex operations:
        1. Converts to lowercase.
        2. Removes bracketed text.
        3. Strips out URLs.
        4. Removes HTML tags.
        5. Eliminates punctuation.
        6. Removes newline characters.
        7. Drops words containing digits.
'''
    text = str(text).lower() # convert text to lowercase
    text = re.sub('\[.*?\]', '', text) 
    text = re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', text) # Remove URl and tags
    text = re.sub('<.*?>+', '', text)
    text = re.sub(r'[^a-z0-9\s]', '', text) # Remove punctuation
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
# Grid Search with Cross-Validation (HalvingGridSearch)

def grid_search(model, parameters):
    """
    Performs a halving grid search over 'parameters' for the given 'model'
    using 5-fold cross-validation. Optimizes the 'f1_weighted' score.
    
    Parameters:
    -----------
    model : estimator
        The base estimator to be optimized (e.g., BernoulliNB()).
    parameters : dict
        Dictionary of hyperparameters to search, e.g., {'alpha': [0.1, 0.5, 1.0]}.

    Returns:
    --------
    grid : HalvingGridSearchCV object
        The fitted HalvingGridSearchCV instance, which contains best_score_ 
        and best_params_ among other attributes.
    """
    
    grid = HalvingGridSearchCV(
        estimator=model,
        param_grid=parameters,
        factor=2,
        cv=5,
        scoring='f1_weighted',
        random_state=42,
        error_score=0
    )
    
    # Fit the grid to the training data
    grid.fit(X_train, y_train)
    
    print("Best Score :", grid.best_score_)
    print("Best Parameters :", grid.best_params_)
    
    # return the fitted grid for access to grid.best_estimator_
    return grid
```

```python
# Predict probabilities
y_score = model.predict_proba(X_test_vect)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curve for each class
for i in range(n_classes):
    plt.figure()
    plt.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f)' % roc_auc[i])
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC for class {i}')
    plt.legend(loc="lower right")
    plt.show()
```
