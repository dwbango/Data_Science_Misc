```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Describe a scenario where the mean and median values of a dataset may differ significantly. 
```

```python
# Right-Skewed Distribution (Mean > Median): 
# Generating sample data: Most bass weights are between 2-4 lbs
weights = np.random.normal(3, 0.5, 200) 

# Adding some trophy bass weights to the data
trophy_bass = np.random.uniform(5, 15, 50)
weights = np.concatenate([weights, trophy_bass])

mean_weight = np.mean(weights)
median_weight = np.median(weights)

# Density Plot
plt.hist(weights, bins=30, edgecolor='black', color='lightgray', density=True, alpha=0.7)
plt.axvline(mean_weight, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median_weight, color='g', linestyle='dotted', linewidth=1, label='Median')
plt.legend()
plt.title('Largemouth Bass Weight Distribution')
plt.xlabel('Weight (lbs)')
plt.ylabel('Density')
plt.grid(axis='y')
plt.tight_layout()
plt.show()


print(f"Mean Weight (Right Skewed): {mean_weight:.2f} lbs")
print(f"Median Weight (Left Skewed): {median_weight:.2f} lbs")
```

```python
# Symmetrical Distribution (Mean = Median)
# Sample data: Most bass weights are between 3-4 lbs with few outliers
weights_symmetric = np.random.normal(3.5, 0.05, 250)

mean_symmetric = np.mean(weights_symmetric)
median_symmetric = np.median(weights_symmetric)

# Visualization
plt.hist(weights_symmetric, bins=30, edgecolor='black', color='lightgray', density=True, alpha=0.5)
plt.axvline(mean_symmetric, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median_symmetric, color='g', linestyle='dotted', linewidth=1, label='Median')
plt.legend()
plt.title('Symmetric Largemouth Bass Weight Distribution')
plt.xlabel('Weight (lbs)')
plt.ylabel('Density')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

print(f"Mean Weight (Symmetric): {mean_symmetric:.2f} lbs")
print(f"Median Weight (Symmetric): {median_symmetric:.2f} lbs")
```

```python
#Left-Skewed Distribution (Mean < Median)
# Bass normally distributed around 4 pounds w/ a 0.5 lb SD
weights_left_skewed = np.random.normal(4, 0.5, 200)

# Adding a few very light bass weights to the data
light_bass = np.random.uniform(0.5, 3.0, 50)
weights_left_skewed = np.concatenate([weights_left_skewed, light_bass])

mean_left_skewed = np.mean(weights_left_skewed)
median_left_skewed = np.median(weights_left_skewed)

# Visualization
plt.hist(weights_left_skewed, bins=30, edgecolor='black', color='lightgray', density=True, alpha=0.7)
plt.axvline(mean_left_skewed, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median_left_skewed, color='g', linestyle='dotted', linewidth=1, label='Median')
plt.legend()
plt.title('Left-Skewed Distribution')
plt.xlabel('Weight (lbs)')
plt.ylabel('Density')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

print(f"Mean Weight (Left-Skewed): {mean_left_skewed:.2f} lbs")
print(f"Median Weight (Left-Skewed): {median_left_skewed:.2f} lbs")
```

```python
#Insights from the Discrepancy:
#Nature of Distribution: If the mean and median are not the same, it's a clear indication that the data distribution is skewed.
#The direction of the skewness (left or right) can be determined by comparing their relative positions.

#Presence of Outliers: A significant difference between the mean and median often suggests the presence of outliers in the dataset.
#For instance, in a right-skewed distribution, the mean will be influenced by high-valued outliers more than the median.

#Data Reliability: The sensitivity of the mean to extreme values can sometimes make it less representative of the central trend of 
#data compared to the median, especially in the presence of outliers.
```

```python
#Solutions to Handle the Discrepancy:
# Use Resilient Statistics: In cases where outliers significantly skew data, consider using resilient statistical measures that may combat outliers. 
#The median is a robust measure of central tendency. Similarly, instead of the range or standard deviation, 
#one could use the interquartile range (IQR) to describe the spread of data. (Shown in my last discussion)

# Data Transformation: Transformations like taking the logarithm can make highly skewed distributions more symmetrical, making the mean more representative.

# Outlier Treatment: If outliers are confirmed as errors or non-representative values, they can be removed or manipulated accordingly.

# Contextual Interpretation: Do not rely on a single measure, analyze both mean and median in context to make informed decisions.
```

```python
#what are some real-world examples of this phenomenon?
#1. Income distribution 
#A commmon scenario of right-skewed data. A majority of people earn average or below-average incomes, and 
#a few people earn extremely high incomes.

# 2. Age at Retirement
# The age that people retire can be left-skewed. Most people retire between 50 and 65, but some might 
# retire early due to various reasons, causing the distribution to skew to the left.

# 3. Examination Scores
# Exam scores can be either left or right-skewed, depending on the difficulty of the exam and the dedication of the students. A very easy exam can
# result in a left-skewed distribution with most students scoring high grades, and a few scoring lower. And vice versa, a very 
# difficult exam can lead to a right-skewed distribution with most students scoring low scores, and a few scoring higher.
```
