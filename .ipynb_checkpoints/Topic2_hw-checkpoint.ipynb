```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df1 = pd.read_csv("downloads/usedcars.csv")
df1['Price'] = df1['Price'] / 100
df1.head()
```

```python
_='''
1. Outlier Identification and Handling: In this task, you will work with a real-world dataset to identify and handle 
outliers. Choose a dataset (e.g., from Kaggle, UCI Machine Learning Repository) from the list of "Repositories for Finding 
Suitable Datasets," located in Class Resources, that exhibits outliers or extreme values. Write a Python script that 
identifies and handles the outliers using at least two methods (e.g., z-score, interquartile range). Use visualization 
techniques to demonstrate the impact of the outliers on measures of central tendency and variability.
'''
```

```python
# Z-Score method to identify outliers

# Calculate z-scores for 'Price'
z_scores = stats.zscore(df1['Price'])

# Identify rows where z-score is greater than 3 or less than -3
outliers_ = (z_scores > 2.5) | (z_scores < - 2.5)

# Filter the dataframe to get only the outlier rows based on z-score method
outliers_z_score = df1[outliers_]
#outliers_z_score
len(outliers_z_score)
```

```python
# IQR Method to identify outliers

# Calculate Q1 and Q3
Q1 = df1['Price'].quantile(0.25)
Q3 = df1['Price'].quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify rows where price is an outlier
outliers_iqr = (df1['Price'] < lower_bound) | (df1['Price'] > upper_bound)

# Filter the dataframe to get only the outlier rows based on IQR method
outliers_iqr_ = df1[outliers_iqr]
# print(outliers_iqr_)
len(outliers_iqr_)
```

```python
#Histogram & Density Plot of price from OG data
plt.figure(figsize=(12, 6))
sns.histplot(df1['Price'], bins=50, kde=True, color='skyblue')
plt.title('Price Distribution: Histogram & Density Plot')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()
```

```python
# Handling the outliers
# The data is right skewed most likely due to luxury cars being included. 
#I am going to remove outlier from the used car data set given the z-score method

# Filter the dataframe to keep only the non-outlier rows based on the z-score method
df_no_outliers = df1[outliers_ == False]

#df_no_outliers = df1[~outliers_] # code above also can be written like this

# outliers_only = df1[outliers_] #outliers only
```

```python
# Calculate and round mean and median for original dataframe
original_mean = round(df1['Price'].mean(), 2)
original_median = round(df1['Price'].median(), 2)

# Calculate and round mean and median for dataframe without outliers
no_outliers_mean = round(df_no_outliers['Price'].mean(), 2)
no_outliers_median = round(df_no_outliers['Price'].median(), 2)

print("Original Data Mean: " + str(original_mean) + ", Median: " + str(original_median))
print("Data Without Outliers Mean: " + str(no_outliers_mean) + ", Median: " + str(no_outliers_median))


# Plotting bar chart

# Set the bar positions
positions = [0, 0.60, 1.50, 2.1] 

# Plotting the bars
plt.figure(figsize=(12, 6))
plt.bar(positions[0], original_mean, color='blue', label='Original Data - Mean', width=0.5)
plt.bar(positions[1], no_outliers_mean, color='green', label='Without Outliers - Mean', width=0.5)
plt.bar(positions[2], original_median, color='blue', label='Original Data - Median', width=0.5, alpha=0.7)  # Reduced alpha for median to differentiate
plt.bar(positions[3], no_outliers_median, color='green', label='Without Outliers - Median', width=0.5, alpha=0.7)  # Reduced alpha for median to differentiate

# Setting labels and title
plt.xticks(positions, ['Original Mean', 'Adjusted Mean', 'Original Median', 'Adjusted Median'])
plt.ylabel('Price')
plt.title('Comparison of Mean and Median Prices')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
_='''
#Bias:
Selection Bias: If the cars in the dataset are not randomly selected, it might not be representative of the general
used car market. For instance, if the dataset has more luxury cars, the average price might be skewed upwards.

# Reporting Bias: If the prices or other features of the cars are self-reported (by sellers), there might be
tendencies to over-report or under-report certain values.

# Confounding Variables:
Location as a Confounder: If the dataset has prices from different cities or regions, 
location can be a confounding variable. Prices might be higher in cities with higher living costs,
not necessarily because of the car's features and milage. Ex: Two identical cars, one in NY and one in NM.

# Impact of confounding variables and Bias on Analysis:
Skewed Results: Bias can skew the results in a particular direction. Selection bias towards luxury cars would make the 
average car price in the dataset higher than in the general market.
Misrepresentation: If certain types of cars (e.g., older models, specific brands) are underrepresented or overrepresented
due to bias, any generalization made based on the dataset would not accurately reflect the broader market.

# Unreliable Conclusions: Reporting bias can lead to unreliable conclusions. If sellers tend to over report the prices 
of their cars, an analysis might suggest that cars are generally more valuable than they actually are in the market.

# False Connections: Without looking at other factors, like where a car is sold, we might think that cars with certain features, 
like sunroofs, are always more expensive. However, it could be that these cars are just more popular in richer areas where all 
cars cost more.

# Confounding variables affect the internal validity of an analysis. Relationships observed in the data 
might not be direct but rather due to some external factor.
'''
```

```python
# Calculate mean and median values of 'Mileage' in df1
mean_mileage = df1['Mileage'].mean()
median_mileage = df1['Mileage'].median()
print(f"Mean Mileage: {mean_mileage}")
print(f"Median Mileage: {median_mileage}")
```

```python
_='''
Right-Skewed Distribution: The difference between the mean and median indicates that the mileage data is right-skewed. 
This means that while most cars have a mileage around or below 80,000, there are some cars with extremely high mileage 
that are affecting the average.

Outliers: The skewness may be due to the presence of outliers, which can be cars with extremely high mileage.

Proposed Solutions:
Remove Outliers: We can identify and remove extreme values using methods like the IQR.
Transform Data: We can change the data's shape using methods like taking the square root or the log. 
This helps make the data more "even," but  changes how we read the data.

I do not see much of a reason to alter the data given the results
'''
```

```python
# Calculate Q1, Q3, and IQR for 'Mileage'
Q1 = df1['Mileage'].quantile(0.25)
Q3 = df1['Mileage'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_filtered = df1[(df1['Mileage'] >= lower_bound) & (df1['Mileage'] <= upper_bound)]

# Calculate the new mean and median for the filtered 'Mileage' data
new_mean_mileage = df_filtered['Mileage'].mean()
new_median_mileage = df_filtered['Mileage'].median()

print(f"New Mean Mileage: {new_mean_mileage}")
print(f"New Median Mileage: {new_median_mileage}")
```

```python
plt.figure(figsize=(10, 6))

# Plot histogram
sns.histplot(df1['Mileage'], bins=50, kde=True, color='skyblue')

# Add lines for mean and median
plt.axvline(mean_mileage, color='r', linestyle='--', label=f'Mean: {mean_mileage:.2f}')
plt.axvline(median_mileage, color='g', linestyle='-', label=f'Median: {median_mileage:.2f}')

# Add legend and labels
plt.legend()
plt.title('Mileage Distribution with Mean and Median')
plt.xlabel('Mileage')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Calculate mode and standard deviation for 'Milage' in df1
mileage_mode = df1['Mileage'].mode()
mileage_sd = df1['Mileage'].std()
count_100000 = (df1['Mileage'] == 100000).sum()

print(f"Mode of Mileage: {mileage_mode}")
print(f"Standard Deviation of Mileage: {mileage_sd:.2f}")
```

```python
_='''
Mean: Gives the average value. It is influenced by all data points, so very high or very low values can skew it. 
It provides insight into the "center" of our data.

Median: The middle value when data is sorted. Not affected by outliers like the mean is, it often gives a 
better "typical" value when data has extremes. It also indicates the "center" of our data.

Mode: The most common value. Useful for seeing what's popular or frequent, but does not tell us much about the 
spread of the data. It primarily informs us about frequency.

Standard Deviation (SD): Measures how spread out the data is. A high SD means values are more spread out; a low SD
means they're closer to the mean. It provides insight into the variability of the data.
'''
```

```python
# Rename the column
df3.rename(columns={" Annual income": "Annual Income"}, inplace=True)

# Convert all values in the 'Annual Income' column to string
df3['Annual Income'] = df3['Annual Income'].astype(str)

# Clean the "Annual Income" column
df3['Annual Income'] = df3['Annual Income'].str.replace('$', '')  # Remove dollar signs
df3['Annual Income'] = df3['Annual Income'].str.replace(',', '')  # Remove commas
df3['Annual Income'] = df3['Annual Income'].str.strip()  # Remove any leading or trailing spaces

# After cleaning, convert back to float
df3['Annual Income'] = df3['Annual Income'].astype(float)

# Calculate measures of central tendency and variability for 'Annual Income'
mean_income = df3['Annual Income'].mean()
median_income = df3['Annual Income'].median()
mode_income = df3['Annual Income'].mode().iloc[0]  # Taking the first mode, if multiple modes exist
range_income = df3['Annual Income'].max() - df3['Annual Income'].min()
variance_income = df3['Annual Income'].var()
sd_income = df3['Annual Income'].std()

print(f"Mean Annual Income: {mean_income:.2f}")
print(f"Median Annual Income: {median_income:.2f}")
print(f"Mode Annual Income: {mode_income:.2f}")
print(f"Range of Annual Income: {range_income:.2f}")
print(f"Variance of Annual Income: {variance_income:.2f}")
print(f"Standard Deviation of Annual Income: {sd_income:.2f}")
```

```python
# Insurance Claim Analysis: Demographic and Health
# https://www.kaggle.com/datasets/thedevastator/insurance-claim-analysis-demographic-and-health/data

df4 = pd.read_csv("downloads/insurance_data.csv")
print(df4.head()) 
print(df4.columns)  # Print column names
```

```python
# 1. Create 'age_group' column using bins
bins = [15, 25, 35, 45, 55, 65]
labels = ["15-24", "25-34", "35-44", "45-54", "55-64"]
df4['age_group'] = pd.cut(df4['age'], bins=bins, labels=labels, right=False)

# 2. Group by 'age_group'
grouped = df4.groupby('age_group')

# 3. Calculate measures of central tendency (mean) and variability (standard deviation) for 'claim'
mean_claims = grouped['claim'].mean()
median_claims = grouped['claim'].median()
std_dev_claims = grouped['claim'].std()

print(mean_claims)
print(median_claims)
print(std_dev_claims)

# 4. Visualize the Results
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_claims.index, y=mean_claims, capsize=10, palette="pastel")
plt.title('Mean Claim Amounts by Age Group with Standard Deviation')
plt.xlabel('Age Group')
plt.ylabel('Mean Claim Amount')
plt.show()

# Interpretation:
# The height of the bars indicates the average claim amount for each age group.
# The error bars show the variability (standard deviation) in claim amounts within each age group.
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="age_group", y="claim", data=df4, palette="pastel")
plt.title('Boxplot of Claim Amounts by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Claim Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for label in labels:
    subset = df4[df4['age_group'] == label]
    sns.kdeplot(subset['claim'], label=label, fill=True)  # Updated here: replaced 'shade' with 'fill'
plt.title('Density Plot of Claim Amounts by Age Group')
plt.xlabel('Claim Amount')
plt.ylabel('Density')
plt.legend(title='Age Group')
plt.show()
```
