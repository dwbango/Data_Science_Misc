## k-means Algorithm and Its Application

The k-means algorithm is an unsupervised clustering technique used to partition data into k distinct clusters based on feature similarity. The primary objective is to minimize the within-cluster sum of squares (WCSS) without aimlessly adding too makny clusters whcih inherently unproductively reduces WCSS. WCSS quantifies the variance within each cluster.

## Steps using the k-means Algorithm:
1. **Initialization**: Select the number of clusters, K, and initialize K centroids from the dataset points.
2. **Assignment**: Assign each data point to the nearest centroid based on the Euclidean distance, forming K clusters.
3. **Update**: Recalculate the centroids as the mean of all points within each cluster.
4. **Iteration**: Repeat the assignment and update processes until the movement of centroids is minimal or reaches a stability criterion, suggesting convergence.
5. Use methods to verify optimal k number of clusters like elbow method and/ or silhouette methods.
6. valiate the model


## Assumptions and Considerations:
- **Spherical Clusters**: Assumes clusters are spherical and often of similar size, which might not hold for complex data structures.
- **Scale Sensitivity**: Performance is influenced by the scale of data features, necessitating standardization or normalization to ensure equal treatment during distance calculation.
- **Initial Centroids**: The algorithm's output can vary significantly with different initial centroids. Multiple runs with different initializations might be necessary to achieve a reliable result.
- **Predefined K**: The number of clusters must be specified beforehand, often determined using methods like the elbow method or silhouette analysis to gauge the optimal cluster count.
- **Outliers**: K-means is sensitive to outliers as they can skew the centroids significantly, leading to less optimal clustering. Outlier detection and removal or robust scaling methods might be needed.
- **Curse of Dimensionality**: High-dimensional data can dilute the meaning of distance measures used in k-means, making clustering less effective. Dimensionality reduction techniques like PCA (Principal Component Analysis) might be employed to mitigate this effect.

In our analysis of the obesity dataset, k-means will be used to uncover distinct patterns and common characteristics among individuals, aiding in the identification of targeted interventions based on shared obesity-related traits.

# Clustering with K-Means
***Finding the Optimal Number of Clusters***

```python
# Conducting elbow method for optimal k
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=31)
    kmeans.fit(obesity_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 15), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

During the application of the elbow method, no distinct elbow point was evident, prompting the utilization of the silhouette method to determine the optimal number of clusters (k) for the k-means model.

```python
# Silhouette method to find optimal amount of clusters 'k'
from sklearn.metrics import silhouette_score

# List to store the silhouette scores
silhouette_scores = []

# Loop over possible values of k
for k in range(2, 26):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=31)
    kmeans.fit(obesity_scaled)
    score = silhouette_score(obesity_scaled, kmeans.labels_)
    silhouette_scores.append(score)
    print("For k={}, Silhouette Score = {:.3f}".format(k, score))

```

```python
# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 26), silhouette_scores, marker='o')
plt.title('Silhouette Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

**Justification for Cluster Selection (k=4):**

Choosing k=4 for the k-means clustering of the obesity dataset is supported by observing a local maximum in the silhouette scores (0.222), which assesses the cohesion and separation of clusters. While the scores are generally low, with k=2 scoring the highest at 0.385, the choice of four clusters provides a detailed segmentation of the population. This level of detail is crucial given the complex and varied factors influencing obesity, making k=4 a strategic balance for actionable insights without excessive model complexity. This number of clusters supports effective public health strategies and personalized interventions, accommodating the multifaceted nature of obesity while avoiding overfitting, typical in higher cluster numbers.

***4.2 Performing K-Means Clustering***

```python
# Using k = 4 for k-means clustering
kmeans = KMeans(n_clusters = 4, init='k-means++', random_state=31)
clusters = kmeans.fit_predict(obesity_scaled)
```

## Validation of the Clustering Model and explanation:

```python
# Use silhouette score to validate the clusters
from sklearn.metrics import silhouette_score
score = silhouette_score(obesity_scaled, clusters)
print("Silhouette Score: ", score)
#what are silhouette scores and how they apply
```

Silhouette Score Analysis: The silhouette score for k=4 was previously determined as 0.222, indicating moderate separation of clusters. This score, while not high, suggests that the clusters are reasonably distinct, though there is room for improvement.

```python
# Dunn index Analysis

distances = euclidean_distances(obesity_scaled)
d_index = dunn(distances, clusters)
print("Dunn Index:", d_index)
```

The Dunn Index for our clustering result is 0.0258, which is relatively low. This suggests that the clusters may not be as well-separated as desired, or that there is considerable variance within some of the clusters. Typically, a higher Dunn Index is preferred as it indicates clearer and more distinct clusters, suggesting that either the number of clusters or the feature selection needs revisiting.
